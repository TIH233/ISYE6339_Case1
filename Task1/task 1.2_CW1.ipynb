```python
import pandas as pd

model_master = pd.read_excel("Model_master.xlsx")
model_master.head()
```

```python
model_master = model_master.rename(columns={
    "Share(MP)": "Share_MP",
    "Price_EUR": "Price_EUR"
})
```

```python
pop_long["key"] = 1
model_master["key"] = 1

metro_sku = pop_long.merge(model_master, on="key").drop("key", axis=1)
```

```python
metro_sku.shape
metro_sku.head(10)
```

```python
metro_sku["Units_Model"] = (
    metro_sku["Total_Units"] * metro_sku["Share_MP"]
).round(0)
```

```python
def fix_rounding(group):
    total_units = group["Total_Units"].iloc[0]
    diff = total_units - group["Units_Model"].sum()

    # Add/subtract diff to the largest-share model (usually F10)
    idx = group["Share_MP"].idxmax()
    group.loc[idx, "Units_Model"] += diff
    return group

metro_sku = metro_sku.groupby(
    ["Countries", "Metropolitan cities", "Year"],
    group_keys=False
).apply(fix_rounding)
```

```python
check = metro_sku.groupby(["Countries","Metropolitan cities","Year"]).agg(
    TotalUnits=("Total_Units","first"),
    SumSKU=("Units_Model","sum")
)
(check["TotalUnits"] - check["SumSKU"]).value_counts().head()
```

```python
metro_sku["Revenue_Model"] = metro_sku["Units_Model"] * metro_sku["Price_EUR"]
```

```python
metro_total_rev = metro_sku.groupby(
    ["Countries", "Metropolitan cities", "Year"],
    as_index=False
)["Revenue_Model"].sum().rename(columns={"Revenue_Model":"Revenue_Total"})
```

```python
metro_sku.to_excel("Task1_2_Metro_SKU_Units_Revenue.xlsx", index=False)
```

```python
with pd.ExcelWriter("Task1_2_Metro_Output.xlsx", engine="openpyxl") as writer:
    pop_long.to_excel(writer, sheet_name="Metro_TotalUnits", index=False)
    metro_sku.to_excel(writer, sheet_name="Metro_SKU", index=False)
    metro_total_rev.to_excel(writer, sheet_name="Metro_TotalRevenue", index=False)
```

```python
import pandas as pd

nonmetro = pd.read_excel(r"Non-Metro_pop.xlsx")
nonmetro.head()
nonmetro.columns
```

```python
# pick population columns (only 2027-2034)
pop_cols = [c for c in nonmetro.columns if c.startswith("Pop_20") and int(c[-4:]) >= 2027]

nonmetro_long = nonmetro.melt(
    id_vars=["Countries"],   # change to "Country" if needed
    value_vars=pop_cols,
    var_name="Year",
    value_name="Population"
)

# Year becomes integer
nonmetro_long["Year"] = nonmetro_long["Year"].str.extract(r"(\d+)").astype(int)

nonmetro_long.head()
```

```python
entry_year_map = {
    "Belgium": 2027, "Germany": 2027, "Luxembourg": 2027, "Netherlands": 2027,
    "Denmark": 2028, "Estonia": 2028, "Finland": 2028, "Latvia": 2028, "Lithuania": 2028,
    "Norway": 2028, "Poland": 2028, "Sweden": 2028,
    "Austria": 2029, "France": 2029, "Ireland": 2029, "Italy": 2029, "Portugal": 2029,
    "Spain": 2029, "Switzerland": 2029,
    "Bulgaria": 2030, "Croatia": 2030, "Cyprus": 2030, "Czech Republic": 2030,
    "Greece": 2030, "Hungary": 2030, "Malta": 2030, "Romania": 2030,
    "Slovakia": 2030, "Slovenia": 2030
}

nonmetro_long["Entry_Year"] = nonmetro_long["Countries"].map(entry_year_map)
```

```python
# MarketYear: 1,2,3,... after entry (clip before entry to 0)
nonmetro_long["MarketYear"] = (nonmetro_long["Year"] - nonmetro_long["Entry_Year"] + 1).clip(lower=0)

def adoption_mp(market_year):
    if market_year <= 0:
        return 0.0
    if market_year == 1:
        return 0.00025
    if market_year == 2:
        return 0.00045
    if market_year == 3:
        return 0.00055
    # year 4+ grows by 0.00005 each year starting from 0.00060
    return 0.00060 + (market_year - 4) * 0.00005

nonmetro_long["AdoptionRate"] = nonmetro_long["MarketYear"].apply(adoption_mp)
```

```python
nonmetro_long["Total_Units"] = (nonmetro_long["Population"] * nonmetro_long["AdoptionRate"]).round(0)
```

```python
model_master = pd.read_excel("Model_master.xlsx")  # same file you used earlier
# must include: Model, Share_MP (or Share(MP)), Price_EUR

# If needed: normalize column names
model_master = model_master.rename(columns={"Share(MP)": "Share_MP"})
```

```python
nonmetro_long["key"] = 1
model_master["key"] = 1

nonmetro_sku = nonmetro_long.merge(model_master, on="key").drop("key", axis=1)
```

```python
nonmetro_sku["Units_Model"] = (nonmetro_sku["Total_Units"] * nonmetro_sku["Share_MP"]).round(0)

def fix_rounding(group):
    total_units = group["Total_Units"].iloc[0]
    diff = total_units - group["Units_Model"].sum()
    idx = group["Share_MP"].idxmax()   # adjust largest share model
    group.loc[idx, "Units_Model"] += diff
    return group

nonmetro_sku = nonmetro_sku.groupby(["Countries","Year"], group_keys=False).apply(fix_rounding)
```

```python
check = nonmetro_sku.groupby(["Countries","Year"]).agg(
    Total=("Total_Units","first"),
    SumSKU=("Units_Model","sum")
)
(check["Total"] - check["SumSKU"]).value_counts().head()
```

```python
nonmetro_sku["Revenue_Model"] = nonmetro_sku["Units_Model"] * nonmetro_sku["Price_EUR"]

nonmetro_total_rev = nonmetro_sku.groupby(["Countries","Year"], as_index=False)["Revenue_Model"].sum()
nonmetro_total_rev = nonmetro_total_rev.rename(columns={"Revenue_Model":"Revenue_Total"})
```

```python
with pd.ExcelWriter("Task1_2_NonMetro_Output.xlsx", engine="openpyxl") as writer:
    nonmetro_long.to_excel(writer, sheet_name="NonMetro_TotalUnits", index=False)
    nonmetro_sku.to_excel(writer, sheet_name="NonMetro_SKU", index=False)
    nonmetro_total_rev.to_excel(writer, sheet_name="NonMetro_TotalRevenue", index=False)
```
